<h1> Data Preparing and Loading</h1>


```python
import torch
import torch.nn as nn # nn contains all the building blocks to create a neural network
import matplotlib.pyplot as plt
import numpy as np

# Create numerical encoding
# Find patterns in the data

# Create a dataset
X = torch.rand(1, 100)
weight=0.7
bias=0.5

Y= weight*X + bias

# Split the dataset (trainig set **, validation set, test set **)

# 80% of the data for training

train_x = X[:, :80] #80% of the data
train_y = Y[:, :80] #80% of the data

# 20% of the data for testing

test_x = X[:, 80:] #20% of the data
test_y = Y[:, 80:] #20% of the data

len(train_x[0]), len(test_x[0])

def plot_graph(train_x, train_y, test_x, test_y, title='None'):
    plt.figure(figsize=(12, 6))
    plt.scatter(train_x[0], train_y[0], label='Training data')
    plt.scatter(test_x[0], test_y[0], label='Testing data')  
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.grid()
    plt.show()
    
plot_graph(train_x, train_y, test_x, test_y, title='Data Distribution')
```

<h1> Build Model </h1>

```python
class LinearRegression(nn.Module):
    # start with random values of weight and bias. look at the training data 
    # and adjust the weight and bias to minimize the loss and closer to ideal values
    # through 2 algorithms: 1.Gradient Descent and 2.Backpropagation
    # for gradient descent requires_grad=True
    # 
    
    def __init__(self):
        super(LinearRegression, self).__init__()
        self.weight = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float32))
        self.bias = nn.Parameter(torch.rand(1, requires_grad=True, dtype=torch.float32))

    def forward(self, x):
        return self.weight * x + self.bias
```
